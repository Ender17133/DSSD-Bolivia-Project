Two scatterplot diagrams were created to show the negative relationship between health care indicators (child mortality & malnutrition) and municipal sustainable development index in Bolivia.
Both scatterplot graphs also show strong negative correlations between healthcare indicators and sustainable development index in Bolivia.

```python
lineplot1 = sb.lmplot(data = healthcare, x = 'sdg2_2_cmc', y = 'imds', line_kws = {"color": "black"}, scatter_kws = {"color": "green", "alpha": 0.5})
plt.xlabel('Chronic malnutrition in children (< 5 years), 2016 (%)', fontsize = 14)
plt.ylabel('Municipal Sustainable Development Index', fontsize = 14)
plt.title('Relationship between Sustainable Development Index and Chronic malnutrition in children in Bolivia', fontsize = 14, fontweight = 'bold')
plt.show()

lineplot2 = sb.lmplot(data = healthcare, x = 'sdg3_2_mrc', y = 'imds', line_kws = {"color": "black"}, scatter_kws = {"color": "green", "alpha": 0.5})
plt.xlabel('Children mortality rate in (< 5 years), 2016 (per. 1,000 live births)', fontsize = 14)
plt.ylabel('Municipal Sustainable Development Index', fontsize = 14)
plt.title('Relationship between Sustainable Development Index and Mortality rate in children in Bolivia', fontsize = 14, fontweight = 'bold')
plt.show()



```

```python
top=healthcare.sort_values(by=["imds"],ascending= False).head(5)
top["Type"] = "High Sustainable Development Index"
```

```python
bottom=healthcare.sort_values(by=["imds"],ascending= False).tail(6)
bottom["Type"] = "Low Sustainable Development Index"
```

```python
dfo=pd.concat([top,bottom])
```

```python
dfo = dfo[(dfo.imds != 36.299999)]
```

Barplot visualization was used to analyze the differences in patters of health variables in 5 cities with the highest levels of sustainable development and in 5 cities with the lowest levels of sustainable development. It was determined that there is a significant difference in instances of chronic malnutration and mortality rates among children between highly developed and least developed cities in Bolivia. 
Cities with the lowest rates of sustainable development have much higher rates of children mortality and chronic malnutrition than cities with the highest sustainable development. 

```python
fig = go.Figure()
fig.add_trace(go.Bar(
    x=dfo['shapeName'],
    y=dfo["sdg2_2_cmc"], 
    name ='Chronic malnutrition in children (< 5 years), 2016 (%)',
))
fig.add_trace(go.Bar(
    x=dfo["shapeName"],
    y=dfo["sdg3_2_mrc"],  
    name ='Children mortality rate in (< 5 years), 2016 (per. 1,000 live births)',
))

fig.update_layout(
                 title_font = dict(family = 'Arial', size = 18), 
                 xaxis_title = '', 
                 legend_title = '', 
                 plot_bgcolor='white', 
                paper_bgcolor='white', 
                 barmode = 'group', 
                 xaxis_tickangle = -45, 
                title = '<b>Rates of Children Chronic Malnutrition and Mortality in cities of Bolivia<b>')
fig.update_xaxes(tickfont = dict(family = 'Arial', size = 16))
fig.update_yaxes(tickfont = dict(family = 'Arial', size = 16))
fig.show()
```
